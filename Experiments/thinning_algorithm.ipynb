```python
%matplotlib inline

from skimage import (
    io,
    util,
    color,
    morphology
)

import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (25, 30)
```

```python
ECG_img = io.imread("./Data/ECG_001_NG.jpeg")
ECG_img = color.rgb2gray(ECG_img)

_ = plt.imshow(
    ECG_img,
    cmap='gray'
)
```

```python
ECG_inv = util.invert(ECG_img)

_ = plt.imshow(
    ECG_inv,
    cmap='gray'
)
```

```python
ECG_thin = morphology.thin(ECG_inv)

_ = plt.imshow(
    ECG_thin,
    cmap='gray'
)
```

```python
ECG_thin = morphology.skeletonize(ECG_inv)

_ = plt.imshow(
    ECG_thin,
    cmap='gray'
)
```
