```python
%matplotlib inline

from skimage import (
    io,
    util,
    color,
    morphology
)

import numpy as np
import pandas as pd
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (25, 30)
```

```python
with open('./Data/ECG_001_NG.pkl', 'rb') as input_file:
    ECG_img = pkl.load(input_file)

_ = plt.imshow(
    ECG_img,
    cmap='gray'
)
```

```python
ECG_thin = morphology.thin(ECG_img)

_ = plt.imshow(
    ECG_thin,
    cmap='gray'
)
```

```python
ECG_skel = morphology.skeletonize(ECG_img)

_ = plt.imshow(
    ECG_thin,
    cmap='gray'
)
```

```python
with open('./Data/ECG_001_TH.pkl', 'wb+') as out_file:
    pkl.dump(ECG_thin, out_file) 
```

```python
with open('./Data/ECG_001_SK.pkl', 'wb+') as out_file:
    pkl.dump(ECG_skel, out_file) 
```
