```python
%matplotlib inline

from skimage import (
    io,
    util,
    color,
    future, 
    feature,
    filters,
    transform,
    segmentation
)

import numpy as np
from scipy import fftpack
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
plt.rcParams['figure.figsize'] = (25, 30)
```

```python
ECG_img = io.imread("./Data/ECG_001.jpeg")

_ = plt.imshow(
    ECG_img
)
```

```python
ECG_data = util.invert(color.rgb2grey(ECG_img))

_ = plt.imshow(
    ECG_data,
    cmap='gray'
)
```

```python
ECG_FFT = np.fft.fft(ECG_data)

_ = plt.imshow(
    np.abs(ECG_FFT), 
    norm=LogNorm(vmin=5)
)
```
